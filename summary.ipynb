```python
import pandas as pd

# i want to compare between column 'Gmeet Avg' and 'Zoom Avg' and 'Meetup Avg' in every row
# and want to know which one is the lowest and highest in every row
# and then i want to add new column named 'Best' and 'Worst' in every row
# 'Best' column is the lowest value between 'Gmeet Avg' and 'Zoom Avg' and 'Meetup Avg'
# fill the 'Best' column with the name of App that has the lowest value
# 'Worst' column is the highest value between 'Gmeet Avg' and 'Zoom Avg' and 'Meetup Avg'

df_client_cpu_ram_1 = pd.read_csv('./summary/case_1_client_cpu_ram_summary.csv')
df_client_cpu_1 = df_client_cpu_ram_1.loc[df_client_cpu_ram_1['Resource'] == 'cpu (%)'].reset_index(drop=True)
df_client_ram_1 = df_client_cpu_ram_1.loc[df_client_cpu_ram_1['Resource'] == 'memory (MB)'].reset_index(drop=True)
df_client_throughput_1 = pd.read_csv('./summary/case_1_client_throughput_summary.csv')
df_client_upload_1 = df_client_throughput_1.loc[df_client_throughput_1['Type'] == 'upload (kbytes)'].reset_index(drop=True)
df_client_download_1 = df_client_throughput_1.loc[df_client_throughput_1['Type'] == 'download (kbytes)'].reset_index(drop=True)

df_client_cpu_1['Gmeet Avg'] = df_client_cpu_1['Gmeet Avg'].astype(float)
df_client_cpu_1['Zoom Avg'] = df_client_cpu_1['Zoom Avg'].astype(float)
df_client_cpu_1['Meetup Avg'] = df_client_cpu_1['Meetup Avg'].astype(float)

df_client_cpu_1['Best'] = df_client_cpu_1[['Gmeet Avg', 'Zoom Avg', 'Meetup Avg']].idxmin(axis=1)
df_client_cpu_1['Worst'] = df_client_cpu_1[['Gmeet Avg', 'Zoom Avg', 'Meetup Avg']].idxmax(axis=1)

df_client_cpu_1['Best'] = df_client_cpu_1['Best'].str.replace(' Avg', '')
df_client_cpu_1['Worst'] = df_client_cpu_1['Worst'].str.replace(' Avg', '')

df_client_ram_1['Gmeet Avg'] = df_client_ram_1['Gmeet Avg'].astype(float)
df_client_ram_1['Zoom Avg'] = df_client_ram_1['Zoom Avg'].astype(float)
df_client_ram_1['Meetup Avg'] = df_client_ram_1['Meetup Avg'].astype(float)

df_client_ram_1['Best'] = df_client_ram_1[['Gmeet Avg', 'Zoom Avg', 'Meetup Avg']].idxmin(axis=1)
df_client_ram_1['Worst'] = df_client_ram_1[['Gmeet Avg', 'Zoom Avg', 'Meetup Avg']].idxmax(axis=1)

df_client_ram_1['Best'] = df_client_ram_1['Best'].str.replace(' Avg', '')
df_client_ram_1['Worst'] = df_client_ram_1['Worst'].str.replace(' Avg', '')

df_client_upload_1['Gmeet Avg'] = df_client_upload_1['Gmeet Avg'].astype(float)
df_client_upload_1['Zoom Avg'] = df_client_upload_1['Zoom Avg'].astype(float)
df_client_upload_1['Meetup Avg'] = df_client_upload_1['Meetup Avg'].astype(float)

df_client_upload_1['Best'] = df_client_upload_1[['Gmeet Avg', 'Zoom Avg', 'Meetup Avg']].idxmin(axis=1)
df_client_upload_1['Worst'] = df_client_upload_1[['Gmeet Avg', 'Zoom Avg', 'Meetup Avg']].idxmax(axis=1)

df_client_upload_1['Best'] = df_client_upload_1['Best'].str.replace(' Avg', '')
df_client_upload_1['Worst'] = df_client_upload_1['Worst'].str.replace(' Avg', '')

df_client_download_1['Gmeet Avg'] = df_client_download_1['Gmeet Avg'].astype(float)
df_client_download_1['Zoom Avg'] = df_client_download_1['Zoom Avg'].astype(float)
df_client_download_1['Meetup Avg'] = df_client_download_1['Meetup Avg'].astype(float)

df_client_download_1['Best'] = df_client_download_1[['Gmeet Avg', 'Zoom Avg', 'Meetup Avg']].idxmin(axis=1)
df_client_download_1['Worst'] = df_client_download_1[['Gmeet Avg', 'Zoom Avg', 'Meetup Avg']].idxmax(axis=1)

df_client_download_1['Best'] = df_client_download_1['Best'].str.replace(' Avg', '')
df_client_download_1['Worst'] = df_client_download_1['Worst'].str.replace(' Avg', '')
```
